```python
import pandas as pd

# Load the dataset with dtype adjustment
df = pd.read_csv("/content/sales_06_FY2020-21 copy.csv", dtype={'order_id': str}, low_memory=False)

# Display basic info
df.info()

# Show first few rows
df.head()
```

To check for missing Values

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]  # Display only columns with missing values
```

To fix data types from floats to strings(str) and Integer(int)

```python
df['order_id'] = df['order_id'].astype(str)
df['cust_id'] = df['cust_id'].astype(str)
df['item_id'] = df['item_id'].astype(str)
df['order_date'] = pd.to_datetime(df['order_date'])  # Convert to datetime
```

To remove duplicate columns

```python
df = df.drop_duplicates()
```

To drop unnecessary columns

```python
df = df.drop(columns=['Phone No.', 'SSN', 'E Mail'])
```

Exploratory Data Analysis (EDA) and Visualization.

Summary statistics

```python
df.describe()
```

Sales Trend Overtime

```python
import matplotlib.pyplot as plt
# Convert order_date to datetime format
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Now extract month and year
df['order_month'] = df['order_date'].dt.to_period('M')

# Extract month and year from order_date
df['order_month'] = df['order_date'].dt.to_period('M')

# Group by month and sum total sales
monthly_sales = df.groupby('order_month')['total'].sum()

# Plot the trend
plt.figure(figsize=(12,6))
monthly_sales.plot(marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

Best Selling Products

```python
# Group by SKU and sum total revenue
top_products = df.groupby('sku')['total'].sum().sort_values(ascending=False).head(10)

# Plot the top products
plt.figure(figsize=(12,6))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product SKU')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

Sales by Region

```python
import seaborn as sns

# Group by region and sum total revenue
region_sales = df.groupby('Region')['total'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='coolwarm')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
